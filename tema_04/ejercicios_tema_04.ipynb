```python
#Ejercicio 15.A
def get_data():
    size=150
    num_tipos=10
    pendiente_base=-8
    np.random.seed(6)
 
    x=np.array([])
    y=np.array([])    
    z=np.array([])
    x_ini=0
    y_ini=50
    z_ini=0
    x_ancho=8
    y_ancho=60
    x_inc=2
    y_inc=y_ancho-3
    z_int=10
 
    for i in range(0,num_tipos):
        pendiente=np.random.uniform(pendiente_base-0.4,pendiente_base+0.4,1)[0]
        x_tipo=np.random.uniform(x_ini,x_ini+x_ancho,size)
        y_tipo=np.random.uniform(y_ini,y_ini+y_ancho,size)+pendiente*x_tipo
        z_tipo=np.full(size, z_ini)
 
        x=np.concatenate((x,x_tipo))
        y=np.concatenate((y,y_tipo))
        z=np.concatenate((z,z_tipo))
 
        x_ini=x_ini+x_inc
        y_ini=y_ini+y_inc
        z_ini=z_ini+z_int
 
    return x,y,z


x,y,z=get_data()

figure=plt.figure(figsize=(10, 8))
axes = figure.add_subplot(1,1,1)
axes.scatter(x,y,c="#82C4F3") 
axes.set_xlabel('Vacuna inyectada (ml)', fontsize=15,labelpad=20,color="#003B80")  
axes.set_ylabel('N Pacientes enfermados cada 100.000 habitantes', fontsize=15,labelpad=20,color="#003B80")
figure.suptitle("Ejercicio 15.A", fontsize=24)
 
```

```python
#Ejercicio 15.B
from sklearn.linear_model import LinearRegression

def get_data():
    size=150
    num_tipos=10
    pendiente_base=-8
    np.random.seed(6)
 
    x=np.array([])
    y=np.array([])    
    z=np.array([])
    x_ini=0
    y_ini=50
    z_ini=0
    x_ancho=8
    y_ancho=60
    x_inc=2
    y_inc=y_ancho-3
    z_int=10
 
    for i in range(0,num_tipos):
        pendiente=np.random.uniform(pendiente_base-0.4,pendiente_base+0.4,1)[0]
        x_tipo=np.random.uniform(x_ini,x_ini+x_ancho,size)
        y_tipo=np.random.uniform(y_ini,y_ini+y_ancho,size)+pendiente*x_tipo
        z_tipo=np.full(size, z_ini)
 
        x=np.concatenate((x,x_tipo))
        y=np.concatenate((y,y_tipo))
        z=np.concatenate((z,z_tipo))
 
        x_ini=x_ini+x_inc
        y_ini=y_ini+y_inc
        z_ini=z_ini+z_int
 
    return x,y,z


x,y,z=get_data()

figure=plt.figure(figsize=(10, 8))
axes = figure.add_subplot(1,1,1)
axes.scatter(x,y,c="#82C4F3") 
axes.set_xlabel('Vacuna inyectada (ml)', fontsize=15,labelpad=20,color="#003B80")  
axes.set_ylabel('N Pacientes enfermados cada 100.000 habitantes', fontsize=15,labelpad=20,color="#003B80")
figure.suptitle("Ejercicio 15.B", fontsize=24)


def get_recta_regresion(x,y):
    model = LinearRegression()
    model.fit(x.reshape(-1, 1), y.reshape(-1, 1))
    x_init=x.min()
    y_init=model.predict([[x_init]])[0,0]
    x_fin=x.max()
    y_fin=model.predict([[x_fin]])[0,0]

    return [x_init,x_fin],[y_init,y_fin]

x_recta_regresion,y_recta_regresion=get_recta_regresion(x,y)

axes.plot(x_recta_regresion,y_recta_regresion,c="#000000")


#La relacin es que cuanta ms vacuna inyectada ms probabilidad hay que enfermar de COVID
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import load_iris
from matplotlib.colors import LightSource
from tabulate import tabulate
from time import perf_counter
from sklearn.datasets import load_breast_cancer

tf.get_logger().setLevel('ERROR')

def plot_metrics(axes,history,title):
    loss_entrenamiento=round(history['loss'][-1],2)
    axes.plot(history['loss'],linestyle="dotted",label="loss entrenamiento:"+ str(loss_entrenamiento))

    axes.set_xlabel('N pocas', fontsize=13,color="#003B80") 
    axes.xaxis.set_major_locator(MaxNLocator(integer=True))

    axes.set_ylabel('Mtricas', fontsize=13,color="#003B80")
    axes.set_ylim(ymin=0,ymax=1.1)
    axes.yaxis.set_major_locator(MultipleLocator(0.1))

    axes.set_title(title)
    axes.set_facecolor("#F0F7FF")
    axes.grid(visible=True, which='major', axis='both',color="#FFFFFF",linewidth=2)
    axes.legend()




def get_datos():
    iris=load_iris()
    
    sepal_length=iris.data[0:99,0]
    petal_length=iris.data[0:99,2]
    flower_type=iris.target[0:99]
    
    x=np.column_stack((sepal_length,petal_length))
    y=flower_type

    return x,y
 
 

 
def compile_fit(capas,epochs,input_dim,x,y):

    np.random.seed(5)
    tf.random.set_seed(5)

    model=Sequential()
    for index,neuronas_capa in enumerate(capas):
        if (index==0):
            model.add(Dense(neuronas_capa, activation='relu',input_dim=input_dim))
        elif (index==len(capas)-1):
            model.add(Dense(neuronas_capa, activation='sigmoid'))          
        else:
            model.add(Dense(neuronas_capa, activation='relu'))

    model.compile(loss='mean_squared_error')
    
    tiempo = perf_counter()
    history=model.fit(x, y,epochs=epochs,verbose=False) 
    tiempo=perf_counter()-tiempo

    return model,tiempo,history
 

def array_to_str(datos):
    s=""
    for index,dato in enumerate(datos):
        if (index==0):
            s=str(dato)
        else:
            s=s+","+str(dato)

    return s

x,y=get_datos()
redes=[  
        [  [4, 8, 4, 2, 1],	    40],  
        [  [8, 16, 8, 4, 1],	40],  
        [  [16, 32, 16, 8, 1],	40], 
        [  [32, 64, 32, 8, 1],	40], 
        [  [64, 128, 64, 8, 1],	40]  
]


input_dim=2


figure=plt.figure(figsize=(20, 11))
figure.suptitle("Redes flores", fontsize=25, color='#003B80')

num_axes=0
for index,datos_red in enumerate(redes):
    capas=datos_red[0]
    epochs=datos_red[1]
    model,tiempo,history=compile_fit(capas,epochs,input_dim,x,y)

    num_axes=num_axes+1
    axes = figure.add_subplot(2,3,num_axes)
    plot_metrics(axes,history.history,"Red:"+array_to_str(capas)) 
figure.tight_layout()

#1 Red: Se debera haber seguido entrenando.
#2 Red: Se debera haber seguido entrenando.
#3 Red: Se debera haber parado sobre la poca 35.
#4 Red: Se debera haber parado sobre la poca 20.
#5 Red: Se debera haber parado sobre la poca 15.

#La mejor red parece la 4 ya que es la ms estable comparada con la 5 que es ms irregular en la 4 poca y en la 34
```
